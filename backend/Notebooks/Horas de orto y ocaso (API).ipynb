Puedes obtener tablas de amanecer y ocaso solar para Barcelona en fechas determinadas usando varias herramientas web y servicios gratuitos:

- Consultar páginas como **salidapuestasol.com** o **es.wschody-zachody.pl**, donde aparecen tablas con las horas diarias de salida y puesta del sol para cada día del año en Barcelona. Por ejemplo, puedes ver filas tipo:

| Día       | Amanecer | Puesta de sol |
|-----------|----------|---------------|
| 22-07-2025| 06:35    | 21:19         |
| 23-07-2025| 06:36    | 21:19         |[1][3]

- También puedes usar calculadoras interactivas, como la de **SunEarthTools** o **PLANETCALC**, donde introduces la ciudad y la fecha, y te genera el horario exacto del amanecer y el ocaso para ese día concreto[6][10].

- Herramientas fotográficas como **PhotoPills** permiten ver, para cualquier ubicación y día, el momento exacto del amanecer y anochecer, y son útiles si requieres precisión para fotografía[2].

- Otras webs como **iCalendario.net** permiten listar y exportar estas horas para cualquier lugar de España y cualquier día del año[8].

En resumen, puedes obtener tablas para fechas específicas consultando webs de efemérides solares o usando calculadoras en línea que generan la información día a día.

Para integrar las horas de amanecer y ocaso de Barcelona en un programa y consultar cualquier día de los últimos 5 años o los 5 próximos, la solución óptima es usar una **API pública** que permita obtener estos datos calculados para cualquier fecha. Destacan:

- La **API Sunrise-Sunset** ofrece un servicio gratuito donde puedes introducir latitud, longitud y la fecha deseada (pasada o futura) y recibirás las horas de amanecer y ocaso en formato JSON. Ejemplo de uso:

  ```
  https://api.sunrise-sunset.org/json?lat=41.3888&lng=2.1590&date=2021-07-27
  ```

  Esto te dará los datos para Barcelona ese día, y puedes automatizarlo para cualquier fecha desde tu programa. *Nota:* las horas vienen en UTC y hay que convertirlas a la zona horaria local y considerar el horario de verano al hacer la conversión[1].

- Alternativamente, puedes consultar la web del **Instituto Geográfico Nacional (IGN)** español, que genera tablas para años completos y provincias pero no ofrece API[5].

- Hay otras webs (como salidapuestasol.com, meteogram.es) que generan tablas anuales, pero no suelen facilitar acceso programático para largos periodos, a diferencia de la API indicada[1][4].

En resumen, la *opción más flexible para integración en software* es usar la API Sunrise-Sunset u otra API similar, consultando la fecha concreta que necesites para cualquier día de los últimos o próximos 5 años, procesando el JSON y aplicando el ajuste horario necesario[1].
